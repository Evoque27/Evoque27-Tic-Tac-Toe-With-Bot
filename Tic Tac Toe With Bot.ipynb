```python
# Tic-Tac-Toe Game with Bot in Python

board=[' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']

# This function inserts the position
def InsertLetter(letter, position):
    board[position] = letter

# This function checks if the position is empty
def CheckPosition(position):
    return board[position] == ' '

# This function draws the game board
def DrawBoard(board):
    print("  "+board[1]+"  |  "+board[2]+"  |  "+board[3]+"  ")
    print("_____|_____|_____")
    print("  "+board[4]+"  |  "+board[5]+"  |  "+board[6]+"  ")
    print("_____|_____|_____")
    print("  "+board[7]+"  |  "+board[8]+"  |  "+board[9]+"  ")
    print("     |     |     ")

# This function checks if anyone has won
def CheckWin(bo, le):
    return (bo[7] == le and bo[8] == le and bo[9] == le) or (bo[4] == le and bo[5] == le and bo[6] == le) or(bo[1] == le and bo[2] == le and bo[3] == le) or(bo[1] == le and bo[4] == le and bo[7] == le) or(bo[2] == le and bo[5] == le and bo[8] == le) or(bo[3] == le and bo[6] == le and bo[9] == le) or(bo[1] == le and bo[5] == le and bo[9] == le) or(bo[3] == le and bo[5] == le and bo[7] == le)

# This function deals with the player's move
def PlayerMove():
    run = True
    while run:
        move = input('Please select a position to place an \'X\' (1-9): ')
        try:
            move = int(move)
            if move > 0 and move < 10:
                if CheckPosition(move):
                    run = False
                    InsertLetter('X', move)
                else:
                    print('Sorry, this space is occupied!')
            else:
                print('Please type a number within the range!')
        except:
            print('Please type a number!')

# This function deals with the computer's move
def ComputerMove():
    possibleMoves = [x for x, letter in enumerate(board) if letter == ' ' and x != 0]
    move = 0

    for let in ['O', 'X']:
        for i in possibleMoves:
            boardCopy = board[:]
            boardCopy[i] = let
            if CheckWin(boardCopy, let):
                move = i
                return move

    cornersOpen = []
    for i in possibleMoves:
        if i in [1,3,7,9]:
            cornersOpen.append(i)
            
    if len(cornersOpen) > 0:
        move = Random(cornersOpen)
        return move

    if 5 in possibleMoves:
        move = 5
        return move

    edgesOpen = []
    for i in possibleMoves:
        if i in [2,4,6,8]:
            edgesOpen.append(i)
            
    if len(edgesOpen) > 0:
        move = Random(edgesOpen)        
    return move

# This function helps the computer to decide the next position
def Random(li):
    import random
    ln = len(li)
    r = random.randrange(0,ln)
    return li[r]
    
# This function checks if the board is full
def BoardFull(board):
    if board.count(' ') > 1:
        return False
    else:
        return True

# Game is started
def Main():
    print('Welcome to Tic Tac Toe!!!')
    print("Player [X] ----- vs ----- Bot [O]")
    print("\n")
    DrawBoard(board)

    while not(BoardFull(board)):
        if not(CheckWin(board, 'O')):
            PlayerMove()
            DrawBoard(board)
        else:
            print('Bot Wins! Better Luck Next Time!')
            break

        if not(CheckWin(board, 'X')):
            move = ComputerMove()
            if move == 0:
                break
            else:
                InsertLetter('O', move)
                print('Computer placed an \'O\' in position', move , ':')
                DrawBoard(board)
        else:
            print('You Win! Good Job!')
            break

    if BoardFull(board):
        print('Tie Game!')
board = [' ' for x in range(10)]
Main()
```
